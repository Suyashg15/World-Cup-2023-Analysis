## Calculating teams performance

```python
team_performance = df.groupby('team').mean()
team_performance.reset_index(inplace=True)
print(team_performance)
```

```python
team_performance.plot(kind='bar', figsize=(15, 9))
plt.title('Team-wise Performance Analysis')
plt.xlabel('Team')
plt.ylabel('Average Performance')
plt.legend(loc='upper right')
plt.show()
```

```python
team_performance['CompositeScore'] = team_performance.sum(axis=1)

ranked_teams = team_performance.sort_values(by='CompositeScore', ascending=False)
```

```python
ranked_teams.head()
```

```python
fig = px.bar(ranked_teams, x='team', y='CompositeScore', color='CompositeScore',
             labels={'CompositeScore': 'Composite Score'},
             title='Team Overall Performance')

fig.update_layout(xaxis=dict(tickangle=45, tickmode='array', tickvals=ranked_teams['team']))
```

### Overall INDIA was the best team to perform throughout the tournament, followed by SOUTH AFRICA and NEW ZEALAND

## Team wise Performance Against Opposition

```python
total_Runs = df[df['bat_or_bowl'] == 'bat'].groupby(['team', 'opposition']).agg({'runs': 'sum'}).reset_index()
total_Wickets = df[df['bat_or_bowl'] == 'bowl'].groupby(['team', 'opposition']).agg({'wickets': 'sum'}).reset_index()
```

```python
Runs_comparison = total_Runs.pivot(index='team', columns='opposition', values='runs')
Wickets_comparison = total_Wickets.pivot(index='team', columns='opposition', values='wickets')
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(Runs_comparison, cmap='Blues', annot=True, fmt='g')
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(Wickets_comparison, cmap='Greens', annot=True, fmt='g')
plt.show()
```

From this stats we can conclude that INDIA scored max no. of runs against NEW ZEALAND folowed by its vise-versa.
Also INDIA took highest wickets against NEW ZEALAND followed by SOUTH AFRICA against AUSTRALIA

## Performance over the time

```python
df['start_date'] = pd.to_datetime(df['start_date'])

fig = go.Figure()

for team in df['team'].unique():
  team_data = df[df['team'] == team]
  battingData = team_data[team_data['bat_or_bowl'] == 'bat'].groupby(['start_date'])[['runs']].sum().reset_index()
  fig.add_trace(go.Scatter(
      x=battingData['start_date'],
      y=battingData['runs'],
      mode='lines',
      name=team
  ))

fig.update_layout(
    title='Teams Performance Trends Over Time through Runs',
    xaxis_title='Start Date',
    yaxis_title='Runs',
    legend=dict(title='Teams'),
    showlegend=True
)

fig.show()
```

```python
df['start_date'] = pd.to_datetime(df['start_date'])

fig = go.Figure()

for team in df['team'].unique():
  team_data = df[df['team'] == team]
  bowlingData = team_data[team_data['bat_or_bowl'] == 'bowl'].groupby(['start_date'])[['wickets']].sum().reset_index()
  fig.add_trace(go.Scatter(
      x=bowlingData['start_date'],
      y=bowlingData['wickets'],
      mode='lines+markers',
      name=team
  ))

fig.update_layout(
    title='Teams Performance Trends Over Time through Wickets',
    xaxis_title='Start Date',
    yaxis_title='Wickets',
    legend=dict(title='Teams'),
    showlegend=True
)

fig.show()
```

From this periodic trend we can conclude that the most consistent Team throughout the tournament was Team INDIA & AUSTRALIA


Overall It can be concluded that INDIA, AUSTRALIA, NEW ZEALAND & SOUTH AFRICA were the best Teams to perform in Cricket World Cup 2023
and AUSTRALIA won the World Cup
